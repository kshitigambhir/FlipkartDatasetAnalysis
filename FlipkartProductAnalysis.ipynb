```python
import pandas as pd

df = pd.read_csv("train.csv")
df.head()
```

```python
# Basic info
df.info()

# Missing values
df.isnull().sum()

# First few rows
df.head()
```

```python
df['Offer %'] = df['Offer %'].str.replace('%', '')
df['Offer %'] = pd.to_numeric(df['Offer %'], errors='coerce')
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df['maincateg'].value_counts()
```

```python
df.groupby('maincateg')['price1'].mean().sort_values(ascending=False)
```

```python
df[['price1', 'Rating']].corr()
```

```python
df.sort_values(by='price1', ascending=False)[['title', 'price1', 'Rating']].head(10)
```

```python
import matplotlib.pyplot as plt

df['maincateg'].value_counts().plot(
    kind='bar',
    color=['#FF6F61', '#6B5B95']  # nice colors
)

plt.title("Number of Products per Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df['price1'], df['Rating'], color='#6B5B95', alpha=0.6)

plt.title("Price vs Rating", fontsize=14)
plt.xlabel("Price")
plt.ylabel("Rating")

plt.grid(alpha=0.3)

plt.show()
```

```python
plt.hist(df['price1'], bins=30, color='#FF6F61', edgecolor='black')

plt.title("Price Distribution", fontsize=14)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
stars = ['star_1f', 'star_2f', 'star_3f', 'star_4f', 'star_5f']

df[stars].sum().plot(
    kind='bar',
    color=['#FF4C4C', '#FF944D', '#FFD700', '#66B2FF', '#4CAF50'],
    edgecolor='black'
)

plt.title("Overall Star Rating Distribution", fontsize=14)
plt.xlabel("Star Rating")
plt.ylabel("Total Count")
plt.show()
```
